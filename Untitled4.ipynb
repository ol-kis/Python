```python
import math
from sympy import*
from sympy.plotting import plot as plt
init_printing()

x = Symbol("x")
fx = -18*x**3 + 5*x**2+10*x - 30
fx
```

```python
p = plt(fx, (x,-7.5,7.5))
```

```python
diff(fx)
```

```python
solve(fx)
```

```python
solve(diff(fx)>0)
```

```python
solve((fx>0))
```
